```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt


# add the project root to the Python path to allow for absolute imports
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
	sys.path.append(project_root)

from mlfromsc.algorithms.polynomial_regression import PolynomialRegression
```

```python
np.random.seed(42)
num_points = 80

X = 6 * np.random.rand(num_points, 1) - 3

y = 1.5 * X**2 - 2 * X + 1 + np.random.randn(num_points, 1) * 2

degree = 2
poly_reg = PolynomialRegression(X, y, degree=degree)

poly_reg.fit(learning_rate=0.01, epochs=50)


plt.figure(figsize=(10, 6))

plt.scatter(X, y, color="blue", alpha=0.7, label="Training Data")

X_new = np.linspace(-3, 3, 100).reshape(-1, 1)
y_new = poly_reg.predict(X_new)

plt.plot(X_new, y_new, color="red", linewidth=3, label="Regression Line")

plt.title(f"2D Polynomial Regression (Degree {degree})")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python

np.random.seed(42)
num_points = 100

x = 2 * np.random.rand(num_points, 1) - 1
y = 2 * np.random.rand(num_points, 1) - 1

z = 1.5 * x + 0.8 * y + 0.5 * x**2 - 0.3 * y**2 + np.random.randn(num_points, 1) * 0.3

X_data = np.hstack((x, y))

degree = 2
poly_reg = PolynomialRegression(X_data, z, degree=degree)

poly_reg.fit(learning_rate=0.1, epochs=200)



fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x, y, z, color="blue", label="Training Data")

x_surf = np.linspace(x.min(), x.max(), 30)
y_surf = np.linspace(y.min(), y.max(), 30)
X_surf, Y_surf = np.meshgrid(x_surf, y_surf)

XY_surf = np.array([X_surf.ravel(), Y_surf.ravel()]).T

Z_surf = poly_reg.predict(XY_surf).reshape(X_surf.shape)

ax.plot_surface(X_surf, Y_surf, Z_surf, color="red", alpha=0.5, label="Regression Surface")

ax.set_title(f"3D Polynomial Regression (Degree {degree})")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

plt.show()
```

```python
np.random.seed(100)
num_points = 200

x = 4 * np.random.rand(num_points, 1) - 2
y = 4 * np.random.rand(num_points, 1) - 2

z = np.sin(x * np.pi / 2) + np.cos(y * np.pi / 2) + x**2 - y**2 + np.random.randn(num_points, 1) * 0.5

X_data = np.hstack((x, y))

degree = 4
poly_reg = PolynomialRegression(X_data, z, degree=degree)

poly_reg.fit(learning_rate=0.01, epochs=500)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x, y, z, color="blue", label="Training Data", alpha=0.6)

x_surf = np.linspace(x.min(), x.max(), 50)
y_surf = np.linspace(y.min(), y.max(), 50)
X_surf, Y_surf = np.meshgrid(x_surf, y_surf)

XY_surf = np.array([X_surf.ravel(), Y_surf.ravel()]).T

Z_surf = poly_reg.predict(XY_surf).reshape(X_surf.shape)

ax.plot_surface(X_surf, Y_surf, Z_surf, color="red", alpha=0.5, label="Regression Surface")

ax.set_title(f"Complex 3D Polynomial Regression (Degree {degree})")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

plt.show()
```
