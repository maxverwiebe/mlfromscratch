```python
import pandas as pd
import sys
import os
# add the project root to the Python path to allow for absolute imports
project_root = os.path.abspath(os.path.join(os.getcwd(), '../..'))
if project_root not in sys.path:
	sys.path.append(project_root)
	
from mlfromsc.utils import PCA
```

```python
data = pd.read_csv(r"/Users/maximilianverwiebe/codingprojects/ml-from-scratch/datasets/diabetes_dataset.csv")

data.head(5)
```

```python
pca = PCA(target_dimensions=3)
```

```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

transformed_data = pca.fit(data)
y = data["Outcome"].values

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(transformed_data.iloc[:, 0], transformed_data.iloc[:, 1], transformed_data.iloc[:, 2], c=y, cmap='viridis')

ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_zlabel("Dimension 3")
ax.set_title("PCA of Diabetes Dataset")

legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)

plt.show()
```
