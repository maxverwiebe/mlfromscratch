```python
import pandas as pd
import sys
import os
import numpy as np

# add the project root to the Python path to allow for absolute imports
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
	sys.path.append(project_root)
```

```python
data = pd.read_csv(r"/Users/maximilianverwiebe/codingprojects/ml-from-scratch/datasets/car_price_prediction.csv")

data.head()
```

```python
from mlfromsc.algorithms.random_forest import RandomForestRegressor
from mlfromsc import TrainTestSplit
```

```python
X = data.drop("Price", axis=1)

y = np.log1p(data["Price"])

X = np.array(X)
y = np.array(y)

train_test_split = TrainTestSplit()
X_train, X_test, y_train, y_test = train_test_split.fit(X, y, test_ratio=0.2, seed=42)
```

```python
rf = RandomForestRegressor(n_trees=100, max_depth=7, min_samples_split=10, max_features=None)
rf.fit(X_train, y_train)
```

```python
from mlfromsc.eval import mean_squared_error, mean_absolute_error, r2_score

y_pred_log = rf.predict(X_test)

# inverse transform the predictions (prices) to the original scale
y_pred = np.expm1(y_pred_log)

y_test_orig = np.expm1(y_test)

mse = mean_squared_error(y_test_orig, y_pred)
mae = mean_absolute_error(y_test_orig, y_pred)
r2 = r2_score(y_test_orig, y_pred)

print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R^2 Score: {r2:.2f}")
```
